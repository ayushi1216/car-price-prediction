**Data Cleaning**

```python
# Year has many non - year values

car['year']
```

```python
car['year'].str.isnumeric().value_counts()
```

```python
car = car[car['year'].str.isnumeric()]
```

```python
car[car['year'].str.isnumeric()].shape
```

```python
# Year obj > Int

car['year'].dtype   #checking type of it
```

```python
car['year'] = car['year'].astype(int)
```

```python
car.info()
```

```python
# Remove "Ask for price" from price

car = car[car['Price'] != "Ask For Price"]
```

```python
# Removing commas of price and converting to int

car['Price'] = car['Price'].str.replace(',', '').astype(int)
```

```python
car.info()
```

```python
car['kms_driven']
```

```python
car['kms_driven'] = car['kms_driven'].str.split(' ').str.get(0).str.replace(',', '')
```

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

```python
# Row with fuel_type 'NaN'
# '~' >> (Tilda) to exclude that row.

car = car[~car['fuel_type'].isna()]
```

```python
# Keep first three words of name

car['name'] = car['name'].str.split(' ').str.slice(0, 3).str.join(' ')
```

```python
#  reset_index to drop the previous wrong ordered index and get ordered new one.
car = car.reset_index(drop=True)
```

```python
car.info()
```

```python
car.describe()
```

```python
car[car['Price'] > 6e6]
```

```python
# Drop rows where car price is greator than 60 Laks

car = car[car['Price'] < 6e6].reset_index(drop=True)
```

```python
# Get this cleaned data in csv
car.to_csv('Cleaned car.csv')
```

**Model**

```python
X = car.drop(columns = 'Price')
y = car['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
```

```python
X.head()
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name', 'company', 'fuel_type']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_), ['name', 'company', 'fuel_type']), remainder = 'passthrough')
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans, lr)
```

```python
pipe.fit(X_train, y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
# Checkingg R2 score

r2_score(y_test, y_pred)
```

```python

# Finding the model with a random state of TrainTestSplit where the model was found to give almost 0.92 as r2_score
#  Every iteration will split the dataset differently into training and test sets.
# random_state = 42 > gives same split everytime.
# random_state = i (data is shuffled differently everytime, giving us 1000 different random splits)
# Maximum R2 score.
scores = []
for i in range(1000):
  X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=i)
  lr = LinearRegression()
  pipe = make_pipeline(column_trans, lr)
  pipe.fit(X_train, y_train)
  y_pred = pipe.predict(X_test)
  scores.append(r2_score(y_test, y_pred))
```

```python
# Check which random_state has highest r2_score
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
  X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=np.argmax(scores))
  lr = LinearRegression()
  pipe = make_pipeline(column_trans, lr)
  pipe.fit(X_train, y_train)
  y_pred = pipe.predict(X_test)
  r2_score(y_test, y_pred)
```
